```python
import pandas as pd
import joblib
import warnings
```

```python
warnings.filterwarnings('ignore', category=UserWarning)
```

```python
class CreditDecisionEngine:
    def __init__(self):
        # Load artifacts generated by Notebooks 1 and 2
        self.model = joblib.load('../models/credit_model.pkl')
        self.scaler = joblib.load('../models/scaler.pkl')
        self.features = self.scaler.feature_names_in_
        
    def evaluate(self, applicant_data):
        df_input = pd.DataFrame([applicant_data])
        
        # Enforce column order
        df_input = df_input[self.features]
        
        # Scale the input and convert it immediately back into a DataFrame with column names
        scaled_array = self.scaler.transform(df_input)
        scaled_df = pd.DataFrame(scaled_array, columns=self.features)
        
        # Predict using the named DataFrame
        prediction = self.model.predict(scaled_df)[0]
        confidence = self.model.predict_proba(scaled_df)[0][1]
        
        return {
            "Creditworthy": bool(prediction == 1),
            "Risk_Score": f"{confidence * 100:.1f}%"
        }

engine = CreditDecisionEngine()
print("Inference Engine initialized and ready. Warnings suppressed.")
```

```python
mock_application = {
    'Account_Balance': 2,
    'Duration_of_Credit_monthly': 24,
    'Payment_Status_of_Previous_Credit': 2,
    'Purpose': 9,
    'Credit_Amount': 1500,
    'Value_Savings_Stocks': 2,
    'Length_of_current_employment': 4,
    'Instalment_per_cent': 2,
    'Sex_Marital_Status': 2,
    'Guarantors': 1,
    'Duration_in_Current_address': 4,
    'Most_valuable_available_asset': 1,
    'Age_years': 28,
    'Concurrent_Credits': 3,
    'Type_of_apartment': 1,
    'No_of_Credits_at_this_Bank': 1,
    'Occupation': 2,
    'No_of_dependents': 1,
    'Telephone': 1,
    'Foreign_Worker': 1
}

result = engine.evaluate(mock_application)
```

```python
print("--- System Output ---")
print(f"Status Approved:  {result['Creditworthy']}")
print(f"Confidence Level: {result['Risk_Score']}")
```
